```python
import numpy as np
import matplotlib.pyplot as plt

# 生成时间序列：模拟一天内（0-23小时）的100个时间点
t = np.linspace(0, 23, 100)

# 生成真实交通流量数据
# 假设基础流量为50辆，加上早高峰（8点左右）和晚高峰（18点左右）的高峰流量
true_flow = 50 \
    + 200 * np.exp(-((t - 8) ** 2) / (2 * 1.5 ** 2)) \
    + 250 * np.exp(-((t - 18) ** 2) / (2 * 1.5 ** 2))
# 为了更真实，加入一定的随机噪声
true_flow += np.random.normal(0, 5, size=t.shape)

# 生成预测流量数据：在真实流量的基础上加入预测误差（均值为0，标准差为10）
pred_flow = true_flow + np.random.normal(0, 10, size=t.shape)

from matplotlib import font_manager
my_font = font_manager.FontProperties(fname='C:\\Windows\\Fonts\\simsun.ttc',size=18) 

# 绘制折线图
# plt.figure(figsize=(10, 6))
plt.figure( figsize=(20, 8), dpi=80 )
plt.plot(t, true_flow, label='真实流量', color='blue', linewidth=1)
plt.plot(t, pred_flow, label='预测流量', color='red', linewidth=1)

plt.xlabel('时间 (小时)',fontproperties=my_font)
plt.ylabel('交通流量 (辆)',fontproperties=my_font)
plt.title('交通流量真实值与预测值对比',fontproperties=my_font)
plt.legend(prop=my_font,loc='best')
plt.grid(alpha=0.4,linestyle='--')
plt.show()
```
